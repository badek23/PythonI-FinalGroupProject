#### *Question 1.1*

```python
#SETUP for later CHECK. Create a mask in which we store the rows that have null Income values
mask = data["Income"].isnull()
data[mask]
```

```python
#CHECKING the mean incomes before insertion first
data.groupby(['Education', 'Marital_Status'])['Income'].mean().round(2)
```

```python
# The inserted income is rounded to 2 decimal points, which is the most reasonable format for financial values
data['Income'] = data['Income'].fillna(data.groupby(['Education', 'Marital_Status'])['Income'].transform('mean')).round(2)
```

```python
#CHECK. Show the same rows that were previously null, and check that their new Income values are the correct ones
data[mask]
```

```python
# CHECKING that there are no longer any nulls in Income
data.isna().sum()
```

#### *Question 1.3*

```python
def get_generation(year):
    if year >= 1981:
        return 'Millenial'
    elif year >= 1965:
        return 'GenX'
    elif year >= 1946:
        return 'BabyBoomer'
    elif year >= 1928:
        return 'Silent'

data['Generation'] = data['Year_Birth'].apply(get_generation)

# We need to exclude the people born before 1928:
data = data[data['Generation'].notnull()]

print(data[['Year_Birth', 'Generation']])
```
