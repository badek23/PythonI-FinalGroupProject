***Question 5***

```python
from math import ceil

# We first need to define the maximum upper income limit for the bins to make sure all income levels are included:
maximum_income_limit = ceil(data["Income"].max() / 10000) * 10000 + 1

bins = np.arange(0, maximum_income_limit, 10000)

averages = data.groupby(pd.cut(data["Income"], bins))[['MntGoldProds','MntWines','NumDealsPurchases']].mean().round(2)
averages
```

```python
# It looks like the dataset contains extreme values, with a lot of rows with null values in between
rows_with_null_averages = averages[['MntGoldProds', 'MntWines', 'NumDealsPurchases']].isnull().all(axis=1).sum()
print(f"There are {rows_with_null_averages} rows with no average values available for all the 3 columns.")
```

```python
# Therefore, we get rid of these rows to avoid having too many rows with null values
cleaned_averages = averages.dropna(subset=['MntGoldProds', 'MntWines', 'NumDealsPurchases'], how='all')
cleaned_averages
```

```python
# Then we directly query the average amount spent on wine for those with an income between 90-100K
amount = cleaned_averages.iloc[9].loc['MntWines']
print(f'The average amount spent on wine for those with an income between 90-100K is {amount}')
```

If the company is not interested in keeping data about all customers with no exceptions, the last row of the averages could also be dropped, since this is a clear outlier:

```python
cleaned_averages_without_outlier = cleaned_averages.iloc[:-1]
cleaned_averages_without_outlier
```
